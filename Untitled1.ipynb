```python
import torch
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt
from pathlib import Path
from glob import glob
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device

device
```

```python
data_path = Path("/kaggle/input/fruit-recognition")
train_path = data_path / 'train' / 'train'
test_path = data_path / 'test' / 'test'
```

```python
import os
def walkthroughdir(path):
    for dir_path, dir_name, filename in os.walk(path):
        print(f"There are {len(filename)} images in {dir_path}")
        

walkthroughdir(train_path)
```

```python
import random
from PIL import Image

train_image_list = sorted(train_path.glob(''))
random_image_path = random.choice(train_image_list)
print(random_image_path)

image_class = random_image_path.parent.stem
#print(image_class)

image = Image.open(random_image_path)
plt.imshow(image)
plt.axis('off')
plt.title(image_class)

# Meta data of image
print(f"The image path is {random_image_path}")
print(f"The class of image is {image_class}")
print(f"The height of image is {image.height}")
print(f"The width of image is {image.width}")
```
